# 1. Temporal Analysis:
   - Identify the frequency of incidents over the years. Are there spikes or trends?

```python
# Resample the data by year and count the number of incidents per year
incidents_per_year = df.resample("Y").size()
incidents_per_year
```

-resample is used to change the frequency of the time series data.Here resampling the data to have a yearly frequency to understand incidents over the years
-size() function counts the number of elements along a particular axis.That is counting the number of incidents that occurred in each year.

Plotting the temporal analysis understands that how nuclear incidents have evolved or changed over the years from 1957 to 2011

```python
import matplotlib.pyplot as plt
```

```python
# Plotting the temporal analysis 
plt.figure(figsize=(10,6))
incidents_per_year.plot(kind = "line",marker="o",color="blue")
plt.title('Number of Nuclear Incidents Over Time (1957-2011)')
plt.xlabel("Year")
plt.ylabel('Number of Incidents')
plt.grid(True)
plt.tight_layout()
plt.xticks(incidents_per_year.index, rotation=45)
plt.show()

#plt.tight_layout() adjusts the layout parameters to minimize overlapping and ensure proper spacing between subplots.
```

Spikes may indicate specific years with a significantly higher number of incidents, while trends may reveal overall changes over time.Here year 1986 have spikes,that is incidents occured more and other years almost follows same pattern of incidents occured

# 2. Geographical Patterns

*   Map the incidents based on their coordinates. Are incidents clustered in specific regions?
*   Analyze if there's any correlation between geographic locations and the type/severity of incidents.


to visualize geographical patterns and identify whether incidents are clustered in specific regions based on their lattitudes and longitudes. 

```python
from mpl_toolkits.basemap import Basemap
```

Basemap module from the mpl_toolkits.basemap library, which was a popular library for creating geographic maps in Matplotlib.

```python
# Get latitude and longitude data
latitude = df["Latitude"].tolist()
longitude = df['Longitude'].tolist()
# tolist() method is often used with NumPy arrays or Pandas Series/DataFrames to convert them to a Python list	Longitude
```

```python
# Create a basic world map using Basemap
plt.figure(figsize=(10, 8))
mapp = Basemap(projection="mill",llcrnrlat=-60,urcrnrlat=90,llcrnrlon=-180,urcrnrlon=180,resolution='c')
mapp.drawcoastlines()
mapp.drawcountries()
mapp.drawmapboundary(fill_color='lightblue')

# Plot incidents on the map
x,y = mapp(longitude, latitude)
mapp.scatter(x, y, marker='o', color='red', s=50)
plt.title('Geographical Distribution of Nuclear Incidents (1957-2011)')
plt.xlabel("LONGITUDE")
plt.ylabel("LATITUDE")
plt.show()
```

    *projection='mill': This sets the map projection to the Miller Cylindrical projection, which is a cylindrical equal-area projection.
    *llcrnrlat=-60: This sets the lower-left corner latitude of the map to -60 degrees.
    *urcrnrlat=90: This sets the upper-right corner latitude of the map to 90 degrees.
    *llcrnrlon=-180: This sets the lower-left corner longitude of the map to -180 degrees.
    *urcrnrlon=180: This sets the upper-right corner longitude of the map to 180 degrees.
    *resolution='c': This sets the resolution of the boundary database to 'c' (crude). Other resolution options include 'l' (low), 'i' (intermediate), 'h' (high), and 'f' (full). The choice of resolution depends on the level of detail you need in your map.

    *drawcoastlines(): used to draw coastlines on a map created,enhancing its visual representation.
    *drawcountries(): used to draw country boundaries on a map created,enhancing its visual representation.
    *drawmapboundary(fill_color=""): used to draw the boundary of the map, including the edge of the map and optionally filling the ocean area with a specified color.

# 4. Casualty Analysis


*   Analyze direct and indirect deaths concerning incident categories.
*   Identify incidents with the highest casualties and their associated categories.



can analyze the direct and indirect deaths associated with each category and identify incidents with the highest casualties

```python
# Convert 'Numbers of Direct Deaths' and 'Numbers of InDirect Deaths' to numeric (some entries have commas)
df['Numbers of Direct Deaths'] = pd.to_numeric(df['Numbers of Direct Deaths'].replace(',', '', regex=True), errors='coerce')
df['Numbers of InDirect Deaths'] = pd.to_numeric(df['Numbers of InDirect Deaths'].replace(',', '', regex=True), errors='coerce')

# Group incidents by Category and sum the direct and indirect deaths
category_casualties = df.groupby('Category')[['Numbers of Direct Deaths', 'Numbers of InDirect Deaths']].sum()

# Display the total casualties per category
print("Casualties based on Incident Category:",category_casualties)



```

    * to_numeric(): used to convert a column or a series to numeric format. This function is particularly useful when you have a column containing strings or other non-numeric values that you want to convert to numeric (integer or float) format.
    * regex=True: This parameter allows you to use regular expressions to clean the data before conversion. If regex=True, then the function treats the provided input as a regular expression and removes characters that do not match the regular expression pattern.
    * errors='coerce': This parameter determines how to handle parsing errors. When errors='coerce', it means that if there are any errors during conversion, the corresponding elements will be set to NaN (Not a Number).

```python
# Plotting the casualties based on incident category
plt.figure(figsize=(14, 8))
category_casualties.plot(kind='bar', stacked=True, color=['skyblue', 'red'])

plt.title('Casualties based on Incident Category')
plt.xlabel('Incident Category')
plt.ylabel('Total Number of Deaths')
plt.legend(title='Casualty Type', labels=['Direct Deaths', 'Indirect Deaths'])
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

# 5. INES Level Analysis:
   - Understand the severity of incidents based INES. Are there commonalities among incidents with high INES levels?
   - Investigate if higher INES levels correlate with specific categories or regions.

    The "INES" stands for the International Nuclear and Radiological Event Scale, and it is used to communicate the safety significance of nuclear and radiological events. 
    Events are categorized from Level 0 (deviation) to Level 7 (major accident).
    
    Level 0 - Deviation:
        No safety significance.
        
    Level 1 - Anomaly:
        Very low safety significance.
        No significant release of radioactive material.
    
    Level 2 - Incident:
        Low safety significance.
        Significant release of radioactive material unlikely.
    
    
    Level 3 - Serious Incident:
        Low safety significance.
        Significant release of radioactive material unlikely.
    
    Level 4 - Accident with Local Consequences:
        Accident with significant safety consequences.
        Limited release of radioactive material beyond the site.
    
    Level 5 - Accident with Wider Consequences:
        Accident with substantial safety consequences.
        Limited release of radioactive material, but with potential wider consequences.
    
    Level 6 - Serious Accident:
        Serious accident.
        Significant release of radioactive material, with potential for broader consequences.
    
    Level 7 - Major Accident:
        Major accident.
        Major release of radioactive material, with widespread and long-term health and environmental effects.

```python
# Group incidents by INES level and Category, then count occurrences
ines_category_counts = df.groupby(['INES\nlevel', 'Category']).size().unstack(fill_value=0)

# Plotting INES level analysis
plt.figure(figsize=(10, 6))
ines_category_counts.plot(kind='bar', stacked=True, cmap='viridis')

plt.title('INES Level Distribution Across Incident Categories')
plt.xlabel('INES Level')
plt.ylabel('Number of Incidents')
plt.legend(title='Incident Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

* stacked=True : 
     useful when you want to show the total value of each category and how it is composed of different subcategories.
* cmap='viridis:
     cmap stands for color map, and it determines the colors used in the plot.'viridis' is one of the built-in colormaps in matplotlib

# 6. Correlation Analysis:
   - Investigate correlations between incident categories, INES levels, geographical locations, and casualty counts.
   - Identify any significant patterns or associations between these variables.

```python
# Select numerical columns for correlation analysis
numerical_columns = ['Numbers of Direct Deaths', 'Numbers of InDirect Deaths', 'INES\nlevel']

# Filter the dataframe for numerical columns
numerical_df = df[numerical_columns]

# Create a correlation matrix
correlation_matrix = numerical_df.corr()

# Display correlation matrix
print("Correlation Matrix:",correlation_matrix)
```

 heatmap is a graphical representation of data where values in a matrix are represented as colors. It's a useful tool for visualizing the relationships between different variables in a correlation matrix. In a heatmap, colors are used to represent the strength and direction of the correlation.

```python
import seaborn as sns
# Plotting the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

plt.title('Correlation Heatmap')
plt.tight_layout()

plt.show()
```

fmt:allows you to control how the numeric values are displayed.
